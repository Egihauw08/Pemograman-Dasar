```python
# Program pengecekan suhu air pendingin

# a. Meminta input dari pengguna
suhu = float(input("Masukkan suhu air pendingin (°C): "))

# b, c, d. Mengecek kondisi suhu dan mencetak status
if suhu <= 90:
    print("STATUS: Normal")
elif suhu < 100:
    print("STATUS: Warning")
else:
    print("ALERT: SHUTDOWN")
```

```python
# Fungsi untuk menghitung efisiensi turbin
def hitung_efisiensi(input_energi, output_energi):
    # b. Mengembalikan nilai efisiensi
    efisiensi = output_energi / input_energi
    return efisiensi

# c. Contoh pemanggilan fungsi
hasil = hitung_efisiensi(500, 400)

# Menampilkan hasil efisiensi
print("Efisiensi turbin =", hasil * 100, "%")
```

```python
# Program klasifikasi beban listrik

# Input
kapasitas_maks = float(input("Masukkan kapasitas maksimum (kW): "))
beban_terpakai = float(input("Masukkan beban terpakai (kW): "))

# Hitung persentase pemakaian
persentase = (beban_terpakai / kapasitas_maks) * 100

# Tentukan status beban
if persentase <= 60:
    status = "Beban Ringan"
elif persentase <= 85:
    status = "Beban Sedang"
else:
    status = "Beban Berat"

# Output
print(f"Persentase pemakaian: {persentase:.2f}%")
print(f"Status beban: {status}")
```

```python
# Data hasil pembacaan suhu pada 5 titik di boiler
suhu_boiler = [92.5, 88.7, 95.1, 102.3, 97.8]

# Looping untuk menampilkan setiap nilai suhu
for i in range(len(suhu_boiler)):
    print("Titik", i + 1, "=", suhu_boiler[i], "°C")

    # Beri peringatan jika suhu melebihi 100°C
    if suhu_boiler[i] > 100:
        print("⚠️ Warning: Suhu di atas 100°C!\n")
```

```python
# Inisialisasi volume awal tangki
tangki = 0

# Loop pengisian bahan bakar
while tangki < 100:
    tangki += 10  # Naikkan 10 liter setiap iterasi
    print("Volume saat ini:", tangki, "liter")

    # Peringatan saat hampir penuh
    if tangki == 80:
        print("WARNING: Hampir penuh!")

    # Pesan saat tangki penuh
    if tangki == 100:
        print("TANGKI PENUH – Pengisian berhenti.")
```

```python
# ==========================================
# PROGRAM MONITORING DAYA & EFISIENSI GENERATOR
# ==========================================

# Fungsi untuk menghitung daya listrik (P = V × I)
def hitung_daya(tegangan, arus):
    return tegangan * arus

# Fungsi untuk menghitung efisiensi generator (%)
def hitung_efisiensi(energi_masuk, energi_keluar):
    return (energi_keluar / energi_masuk) * 100

# Fungsi utama (main program)
def main():
    print("=== SISTEM MONITORING ENERGI GENERATOR ===\n")

    # Perulangan while untuk membaca beberapa unit generator
    unit = 1
    while unit <= 3:  # misalnya ada 3 generator
        print(f"\n--- Generator {unit} ---")

        # Input data
        tegangan = float(input("Masukkan tegangan (Volt): "))
        arus = float(input("Masukkan arus (Ampere): "))
        energi_masuk = float(input("Masukkan energi masuk (kJ): "))
        energi_keluar = float(input("Masukkan energi keluar (kJ): "))

        # Hitung daya dan efisiensi menggunakan fungsi
        daya = hitung_daya(tegangan, arus)
        efisiensi = hitung_efisiensi(energi_masuk, energi_keluar)

        # Tampilkan hasil
        print(f"\nDaya listrik: {daya} Watt")
        print(f"Efisiensi: {efisiensi:.2f}%")

        # Percabangan if-elif-else untuk menilai kondisi generator
        if efisiensi >= 90:
            print("STATUS: Generator sangat efisien ✅")
        elif efisiensi >= 75:
            print("STATUS: Generator efisien ⚙️")
        elif efisiensi >= 60:
            print("STATUS: Generator kurang efisien ⚠️")
        else:
            print("STATUS: Generator tidak efisien ❌")

        unit += 1  # lanjut ke unit berikutnya

    # Setelah semua generator dihitung, tampilkan ringkasan
    print("\n=== PEMERIKSAAN SUHU OPERASI ===")
    suhu_operasi = [85, 92, 97, 105, 89]  # contoh data suhu
    for i in range(len(suhu_operasi)):
        print(f"Unit {i+1}: {suhu_operasi[i]}°C", end="  ")
        if suhu_operasi[i] > 100:
            print("⚠️ WARNING: Suhu berlebih!")
        else:
            print("✅ Suhu normal.")

    print("\n=== PENGISIAN TANGKI BAHAN BAKAR ===")
    volume = 0
    while volume < 100:
        volume += 20
        print(f"Tangki terisi {volume} liter")
        if volume == 80:
            print("WARNING: Hampir penuh!")
        elif volume >= 100:
            print("TANGKI PENUH – Pengisian berhenti.")
            break

# Jalankan program utama
main()
```
